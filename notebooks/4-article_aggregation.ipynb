```python
import pandas as pd
from tqdm import tqdm
```

```python
datasets = []
```

```python
import os
for file in os.listdir("../data/processed/articles/"):
    if file.endswith(".csv"):
        datasets.append(os.path.join("../data/processed/articles/", file))
```

```python
FULLFRAME = pd.read_csv(datasets[0])
FULLFRAME = FULLFRAME[FULLFRAME.article_author!='[]']
FULLFRAME
```

```python
for datafile in tqdm(datasets[1:]):
    df = pd.read_csv(datafile)
    
    if 'article_author' in df.columns:
        df_filter = df[df.article_author!='[]']
        FULLFRAME = pd.concat([FULLFRAME,df_filter])
    
    
    

```

```python
for column_name in FULLFRAME.columns:
    FULLFRAME = FULLFRAME.rename(columns={column_name:'art_'+column_name})
```
